## 利用`plt.scatter`画散点图

`plt.scatter`是一个可以创建散点图的函数

````python
plt.scatter(x, y, marker = 'o')
````

### `plt.scatter`和`plt.plot`创建散点图的区别

`plt.scatter`创建散点图时具有更高的灵活性, 可以**单独控制每个散点与数据匹配**, 也可以让每个散点具有不同的属性

在面对大型数据集的时候, `plt.plot`效率高于`plt.scatter`, 因为plt.scatter需要对每个散点分别渲染, 但是plt.plot散点基本被复制

````python
rng = np.random.RandomState(10)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000  * rng.rand(100)
plt.scatter(x, y, c = colors, s = sizes, alpha = 0.3, cmap = 'hsv')  # cmap: 颜色图谱
````

cmap: 颜色图谱（colormap), 默认绘制为RGB(A)颜色空间。
