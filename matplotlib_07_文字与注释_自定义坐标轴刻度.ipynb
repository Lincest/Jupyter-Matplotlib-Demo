## 1 | 例子-美国出生人数

```python
births = pd.read_csv('births.csv')
```

```python
# np.percentile用于计算百分位数, 见: https://blog.csdn.net/u011630575/article/details/79451357
# 第20, 50, 75分位数
quartiles = np.percentile(births['births'], [20, 50, 75])
quartiles
```

```python
# 看不懂...但是这些都不是重点, 先不深究了
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query('(births > @mu - 5 * @sig) & (births < @mu + 5 * @sig)')
births['day'] = births['day'].astype(int)
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, format = '%Y%m%d')
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.index = [pd.datetime(2012, month, day) for (month, day) in births_by_date.index]
```

```python
fig, ax = plt.subplots(figsize = (12, 4))
births_by_date.plot(ax = ax)
```

```python
# 在图上增加文字标签
style = dict(size = 10, color = 'gray')

ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha = 'center', **style) # (x坐标, y坐标, 字符串, 可选参数), 这里ha表示horizonal alignment即水平对齐方式
# 用法可参考: https://blog.csdn.net/TeFuirnever/article/details/88947248
fig
```

```python
# 设置坐标轴标题
ax.set(title = "USA births by day of year(1969 - 1988)",
      ylabel = "average daily births",
      xlabel = "month")
fig
```

```python
# 设置x轴刻度值, 让月份居中显示
# 下面都是啥玩意, 书上也不解释下
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday = 15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'));
fig
```

## 2 | 坐标变换与文字位置

三种预定义的变换方式: 

`ax.transData`: 以数据为基准的坐标变换 (默认的) 以x, y轴标签作为数据坐标

`ax.transAxes`: 以坐标轴为基准的坐标变换  以坐标轴尺寸比例作为坐标(下图的白色区域)

`fig.transFigure`: 以图形为基准的坐标变换  以图形尺寸比例为基准呈现坐标(下图的灰色区域)

```python
fig, ax = plt.subplots(facecolor = 'lightgray')
ax.axis([0, 10, 0, 10])

ax.text(1, 5, '. Data', transform = ax.transData)
ax.text(0.5, 0.1, '. Axes', transform = ax.transAxes)
ax.text(0.2, 0.2, '. Figure', transform = fig.transFigure)
```

需要注意, 对应的坐标位置都是字符串的左部, 也就是图中的`.`位置

## 4 |  自定义坐标轴刻度

### 4-1 | 主要刻度和次要刻度

每个坐标轴都有主要刻度线(major)和次要刻度线(minor), 主要刻度往往更大或更显著, 而次要刻度往往更小, 一般情况下不会用到次要刻度, 对数图等会用到次要刻度

通过设置每个坐标轴的`formatter`和`locator`对象, 可自定义这些刻度属性

```python
print(ax.xaxis.get_major_locator())
print(ax.xaxis.get_minor_locator())
print(ax.xaxis.get_major_formatter())
print(ax.xaxis.get_minor_formatter())
```

#### 隐藏刻度标签

通过`plt.NullLocator`和`plt.NullFormatter`实现


```python
ax = plt.axes()
ax.plot(np.random.rand(50))
# y轴隐藏刻度和标签; x轴隐藏标签,保留刻度
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
fig, ax = plt.subplots(5, 5, figsize = (5, 5))
fig.subplots_adjust(hspace = 0, wspace = 0)
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range(5):
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[11*i + j], cmap = "bone")
```

### 4-2 | 增减刻度数量

```python
fig, ax = plt.subplots(4, 4, sharex = True, sharey = True)
```

```python
# 利用plt.MaxNLocator()设置最多需要多少显示刻度
for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(plt.MaxNLocator(2))
fig
```

### 4-3 | 花哨的刻度格式

可以通过`MultipleLocator`实现

比如我们希望正弦函数的刻度均为$\pi$的整数倍

```python
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw = 3, label = 'Sine')
ax.plot(x, np.cos(x), lw = 3, label = 'Cosine')

ax.grid(True)
ax.legend(frameon = False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi)
```

```python
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
fig
```

利用自定义函数结合`plt.FuncFormatter`实现刻度上数学符号的显示, 比如上面我们不希望显示小数, 而希望显示$\frac{\pi}{2}$这样的数

```python
def format_func(value, tick_number):
    # 找到pi/2的倍数
    N = int(np.round(2 * value / np.pi))
    if (N == 0):
        return "0"
    elif N == 1:
        return r"$\pi / 2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi / 2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
```

```python
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```
