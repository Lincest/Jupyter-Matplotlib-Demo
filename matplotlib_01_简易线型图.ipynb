```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

将图形保存为文件:

`fig.savefig('my_figure.png')`

**matlab风格的接口**

```python
plt.subplot(2, 1, 1)  # (行, 列, 子图编号)
plt.plot(x, np.sin(x))
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

**面向对象接口**

面向对象接口可以适应更加复杂的场景, 更好地控制图形

```python
# ax是包含两个Axes对象的数组
fig, ax = plt.subplots(2)

# 在每个对象上调用plot方法
ax[0].plot(x, np.sin(x))

ax[1].plot(x, np.cos(x))
```

## 简易线型图

在matplotlib中, figure可以被看成是一个可以容纳各种坐标轴, 图形, 文字和标签的容器

axes是一个**带有刻度和标签的矩形**, 最终会包含所有可视化的图形元素

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

```python
# 或者用plt
plt.plot(x, np.sin(x))
```

```python
# 如果想在一张图中创建多条线, 可以重复调用plot命令
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### 调整图形: 线条的颜色和风格

```python
plt.subplot(5, 1, 1)
plt.plot(x, np.sin(x - 0), color = 'blue')  # 颜色名称
plt.subplot(5, 1, 2)
plt.plot(x, np.sin(x - 1), color = 'g')  # 颜色代码
plt.subplot(5, 1, 3)
plt.plot(x, np.sin(x - 2), color = '0.75')  # 范围在0-1的灰度值
plt.subplot(5, 1, 4)
plt.plot(x, np.sin(x - 3), color = '#FFDD44')  # 十六进制RGB
plt.subplot(5, 1, 5)
plt.plot(x, np.sin(x - 4), color = (1, 0.2, 0.3) ) # 0-1RGB元组
```

```python
# 可以将linestyle和color编码结合起来作为plt.plot函数的一个非关键字参数使用
# 例如:
plt.plot(x, x + 1, '--c')  # 青色虚线
```

### 调整图形: 坐标轴上下限

第一种方法: 通过xlim和ylim

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

第二种方法: 通过传入`[xmin, xmax, ymin, ymax]`, 利用`plt.axis()`

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 10, -1.5, 1.5])
```

```python
# 还可以自动收紧坐标轴, 不留空白区域
plt.plot(x, np.sin(x))
plt.axis('tight')
```

### 设置图形标签

```python
plt.plot(x, np.sin(x))
# 标题
plt.title("A single curve")
# x轴标签
plt.xlabel("x")
# y轴标签
plt.ylabel("I am y")
```

```python
# legend设置标签
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.legend()
```

### plt和ax方法

虽然绝大多数的plt函数都可以直接转换成ax方法, 但是并非所有的命令都可以这样用, 尤其是用来设置坐标轴上下限, 坐标轴标题和图形标题的函数

`plt.xlabel` -> `ax.set_xlabel` (ylabel同)

`plt.xlim` -> `ax.set_xlim` (ylim同)

`plt.title` -> `ax.set_title`

可以通过`ax.set()`方法一次性设置所有的属性

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim = (0, 10), ylim = (-2, 2), xlabel = 'x', ylabel = 'sin(x)', title = "A Simple Plot")
```
