```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rc('font', family='SimHei', size=13)
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

## 1 | 例子: 美化直方图

```python
plt.hist(np.random.randn(1000))
```

```python
# 手动调整
# 1. 灰色背景
ax = plt.axes(fc = '#E6E6E6')
ax.set_axisbelow(True)

# 2. 白色网格线
plt.grid(color = 'w', linestyle = 'solid')

# 3. 隐藏坐标轴的线条
for spine in ax.spines.values():
    spine.set_visible(False)

# 4. 隐藏上边和右边的刻度
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# 5. 弱化刻度与标签
ax.tick_params(colors = 'gray', direction = 'out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
# 6. 设置直方图的轮廓色和填充色
x = np.random.randn(1000)
ax.hist(x, edgecolor = '#E6E6E6', color = '#EE6666')
```

## 2 | 修改默认配置: `rcParams`

在matplotlib每次加载时, 都会定义一个运行时配置rc, 其中包含了所有你创建的图形元素的默认风格, 可以用`plt.rc`简便方法随时修改这个配置

```python
from matplotlib import cycler
colors = cycler('color', ['#EE6666', '#3388BB', '#9988DD', '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor = '#E6E6E6', edgecolor = 'none', axisbelow = 'True', grid = 'True', prop_cycle = colors)
plt.rc('grid', color = 'w', linestyle = 'solid')
plt.rc('xtick', direction = 'out', color = 'gray')
plt.rc('ytick', direction = 'out', color = 'gray')
plt.rc('patch', edgecolor = '#E6E6E6')
plt.rc('lines', linewidth = 2)
```

## 3 | 样式表

使用某种风格: `plt.style.use('stylename')`

临时更换风格: `with plt.style.context('stylename'): `

```python
# 查看所有风格
plt.style.available
```

```python
# 绘制基本图形, 用于下面对样式的演示
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize = (11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc = 'lower left')
```

```python
# 重置为默认风格
plt.rcParams.update(mpl.rcParamsDefault)
hist_and_lines()
```

```python
# 个人感觉比较好看的: bmh, ggplot, seaborn
for stylei in plt.style.available:
    with plt.style.context(stylei):
        hist_and_lines()
        plt.title(stylei)
```
