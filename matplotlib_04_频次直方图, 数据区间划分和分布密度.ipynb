```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

## 1 | `plt.hist`直接创建频次直方图

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
# 个性化选项
# 参数解释: https://blog.csdn.net/qq_37007384/article/details/102880242
# bins划分间隔
# density的类型是 bool型，指定为True,则为频率直方图，False，频数直方图
plt.hist(data , bins = 30, density = True, alpha = 0.5, histtype = 'stepfilled', color = 'steelblue', edgecolor = 'none')
```

### 同坐标轴的多个频次直方图

利用`histtype = 'stepfilled'`和`alpha`搭配使用

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype = 'stepfilled', alpha = 0.3, density = True, bins = 40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

## 2 | 二维频次直方图与数据区间划分

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

`plt.hist2d`二维频次直方图

```python
plt.hist2d(x, y, bins = 30, cmap = 'Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

`plt.hexbin`六边形区间划分


```python
plt.hexbin(x, y, gridsize = 30, cmap = 'Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```
